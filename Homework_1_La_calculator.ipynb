```python

#https://developers.google.com/gmail/api/quickstart/python
#https://developers.google.com/gmail/api/guides/sending


# How to use it: 
#follow this steps: https://chozinthet20602.medium.com/sending-email-with-python-using-gmail-api-33628e36306a
# then the first run must be on-premise, because the Virtual machine cant open a browser to retrieve the token.json file. 
# When you have the token.json file, You copy to the Virtual machine session, an now run this cell.
# Enjoy: Now you can send emails.


from __future__ import print_function
import base64
import os.path
from email.mime.text import MIMEText
from googleapiclient.discovery import build
from google_auth_oauthlib.flow import InstalledAppFlow
from google.auth.transport.requests import Request
from google.oauth2.credentials import Credentials

# If modifying these scopes, delete the file token.json.
SCOPES = ['https://mail.google.com/']

creds = None
if os.path.exists('token.json'):
    creds = Credentials.from_authorized_user_file('token.json', SCOPES)
# If there are no (valid) credentials available, let the user log in.
if not creds or not creds.valid:
    if creds and creds.expired and creds.refresh_token:
        creds.refresh(Request())
    else:
        flow = InstalledAppFlow.from_client_secrets_file('credentials.json', SCOPES)
        creds = flow.run_local_server(port=0)
    # Save the credentials for the next run
    with open('token.json', 'w') as token:
        token.write(creds.to_json())


service = build('gmail', 'v1', credentials=creds)


def create_message(sender, to, subject, message_text):
    message = MIMEText(message_text)
    message['to'] = to
    message['from'] = sender
    message['subject'] = subject
    return {'raw': base64.urlsafe_b64encode(message.as_string().encode()).decode()}


def send_message(service, user_id, message):
    try:
        message = (service.users().messages().send(userId=user_id, body=message)
                   .execute())
        print('Message Id: %s' % message['id'])
        return message
    except Exception as error:
        print(error)
```

```python
# imports 

import pdb
from datetime import datetime as dt
import requests

# error handler

def get_input(string=None):

    if string:
        print(string)

    try:
        choice = int(input())

        if type(choice) != type(1):
            print(f"Please type number inputs, you typed {choice}\n\n")
            choice = None

    except Exception as e:
        print(f"error {e}\n")
        print("restart calculator \n")
        choice = None

    return choice



# mail sender: 

def send_result(title, message, recipient=None, option=1):

    if  option==1 or (not recipient):

        URL_ERROR_WEBHOOK = "https://teams.webhook.com" # replace with the correct URL

        json={
        "summary": title,
        "sections": [{
            "activityTitle": title,
            "activitySubtitle": message
            }],
        }

        response = requests.post(
        url=URL_ERROR_WEBHOOK,
        headers={"Content-Type": "application/json"},
        json=json,
        )

        print(response.status_code)   # Should be 200

    if option==2 and recipient: 
        email = create_message('me', recipient, title, message)
        try:
            print(send_message(service=service, user_id='me', message=email))   
            print("Success..")   

        except Exception as e:
            print("ERROR:", e)


# funciones de interfaz 
def interfaz():

    not_close = True

    while not_close:
        print()
        print("bienvenido a la calculator")
        print("elija una opcion: ")
        print("opcion 1: sumar")
        print("opcion 2: multiplicar o dividir")             
        print("opcion 3: potencia")
        print("SALIR = 4")
        print()

        choice = get_input()
        
        if choice == 4:
            break

        if choice == 2:
            dividir = get_input("opcion 1: multiplicar \nopcion 2: dividir \n")             


        a = get_input("ingrese primer elemento ")
    
        b = get_input("ingrese segundo elemento ")


        if choice == 1:
            resultado = sumar(a,b)

        elif choice == 2:

            resultado = multiplicar(a,b,dividir)

        elif choice == 3:
            resultado = potencia(a,b)


        my_time = dt.now()

        message = f'####   su resultado es: {resultado}, Date: {my_time.strftime("%-d %B %Y, %I:%M:%S%p")} #####'
        title = "last calculus"

        print(message)

        email = input("\nType the Email to send results \n")             

        send_result(title, message)
        send_result(title, message, email, option=2)
        



# funciones de calculo

def sumar(a,b):

    #d = eval(a)

    c = a + b

    return c

def multiplicar(a,b, dividir): 

    if dividir == 2:
        try:
            c = a / b

        except Exception as e: 
            print(f"ERROR, zero division?. Details {e}\n\n")
            c = e
    
    else:
         c = a * b       

    return c

def potencia(a,b):

    c = a**b

    return c

```
